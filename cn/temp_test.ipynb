```python
import json
import tiktoken
from pathlib import Path

def split_jsonl(
    input_path,
    train_path,
    val_path,
    max_length,
    train_ratio=0.9,
    encoding_name="cl100k_base"
):
    tokenizer = tiktoken.get_encoding(encoding_name)

    # ---------- 第一遍：统计合法样本数 ----------
    total_valid = 0
    with open(input_path, "r", encoding="utf-8") as f:
        for line in f:
            obj = json.loads(line)
            text = obj.get("text", "")
            if len(tokenizer.encode(text)) >= max_length:
                total_valid += 1

    train_cutoff = int(total_valid * train_ratio)
    print(f"Total valid samples: {total_valid}")
    print(f"Train samples: {train_cutoff}")
    print(f"Val samples: {total_valid - train_cutoff}")

    # ---------- 第二遍：真正写文件 ----------
    cur_idx = 0
    with open(input_path, "r", encoding="utf-8") as fin, \
         open(train_path, "w", encoding="utf-8") as ftrain, \
         open(val_path, "w", encoding="utf-8") as fval:

        for line in fin:
            obj = json.loads(line)
            text = obj.get("text", "")
            if len(tokenizer.encode(text)) <= max_length:
                continue

            if cur_idx < train_cutoff:
                ftrain.write(json.dumps(obj, ensure_ascii=False) + "\n")
            else:
                fval.write(json.dumps(obj, ensure_ascii=False) + "\n")

            cur_idx += 1

    print("Done ✔")


split_jsonl(
    input_path="part.jsonl",
    train_path="train.jsonl",
    val_path="val.jsonl",
    max_length=256
)
```
